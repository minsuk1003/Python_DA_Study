```python
import pandas as pd

# 데이터프레임 불러오기
latest_matches = pd.read_csv("../Week 2/spi_matches_latest.csv")
```

```python
# DB로 저장
con = sqlite3.connect("C:/Users/kms10/OneDrive - 경희대학교/파이썬 스터디/Python_DA_Study/Week 3/kospi.db")
latest_matches.to_sql("latest_matches", con) # latest_matches를 TABLE_NAME으로 지정
```

```python
# sqlite에서 데이터프레임으로 불러오기
df = pd.read_sql("SELECT * FROM latest_matches", con, index_col='index')
df
```
