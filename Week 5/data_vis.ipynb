### 1. 원하는 해외주식 4종목의 2023년 주가를 불러오고, 4개 종목의 시간대별 종가를 시각화하세요.

조건 : 최대한 이행하세요

1. 한 종목당 하나의 그래프로 시각화하고, **2행 2열로 총 4개의 그래프를 그리도록 설정**
2. 각 그래프의 제목은 종목명, x축 라벨은 "Date", y축 라벨은 "Closing Price"로 설정
3. 그리드 추가
4. 글자가 겹치지 않도록 그래프 크기 확대
5. 각 종목별로 어울리는 색을 직접 지정 (ex. 애플은 검정, 넷플릭스는 빨강 등)
6. 최종 이미지 저장

```python
import FinanceDataReader as fdr
import matplotlib.pyplot as plt

# 종목 리스트와 종목명을 딕셔너리로 정의
stocks = {'AAPL': 'Apple', 'NVDA': 'Nvidia', 'TSLA': 'Tesla', 'NFLX': 'Netflix'}

# 직접 지정한 색상 리스트
custom_colors = ['#999999', 'limegreen', '#CC0000', '#E50914']

# stocks.items()와 custom_colors의 길이를 맞춤
stocks_and_colors = list(zip(stocks.items(), custom_colors))

# Figure 객체 생성 및 크기 지정
fig, axes = plt.subplots(2, 2, figsize=(15,10))

# 서브플롯에 그리드와 데이터 플로팅
for (stock_info, color), ax in zip(stocks_and_colors, axes.flatten()):
    stock_code, stock_name = stock_info
    stock_data = fdr.DataReader(stock_code, '2023')['Close']
    ax.plot(stock_data, label=stock_name, linewidth=2, color=color)
    ax.set_title(stock_name)
    ax.set_xlabel('Date')
    ax.set_ylabel('Closing Price')
    ax.grid(True)

# tight_layout 함수 호출
plt.tight_layout()

# 생성한 Figure 객체의 크기를 지정하여 출력
plt.savefig("stocks.png", dpi=200)
```

### 2. 음악 장르(playlist_genre)별 인기도(track_popularity) 분포를 시각화하세요.

조건

1. 한 그래프에 모든 장르의 분포가 나타날 수 있도록 설정
2. 제목은 "음악 장르별 인기도 분포"로 설정
3. x축 라벨은 "장르", y축 라벨은 "인기도"로 설정
4. 그래프 크기는 (12,6)으로 설정
5. 장르별로 색을 직접 지정
    - pop : 밝은 계열, rap : 회색, rock : 빨강, latin : 초록, r&b : 보라, edm : 파랑
6. 최종 이미지 저장

```python
# 장르별 색상 지정
genre_palette = {
    'pop': 'skyblue',
    'rap': 'gray',
    'rock': 'red',
    'latin': 'green',
    'r&b': 'purple',
    'edm': 'blue',
}

# 그래프 크기 지정
plt.figure(figsize=(12, 6))

# 박스플롯 그리기
sns.boxplot(x="playlist_genre", y="track_popularity", data=df, palette=genre_palette)

# 그리드 추가
plt.grid(True, linestyle="--", alpha=0.7)

# 그래프 제목과 축 레이블 추가
plt.title("음악 장르별 인기도 분포")
plt.xlabel("장르")
plt.ylabel("인기도")

# 이미지로 저장
plt.savefig("spotify_with_grid.png")

# 그래프 보여주기
plt.show()
```
